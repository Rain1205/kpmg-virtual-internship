```python
from datetime import date

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

# apply 函数
df_Merge_Transactions_Customer_Demographic ["Age"] = df_Merge_Transactions_Customer_Demographic ["DOB"].apply(calculate_age)
```
